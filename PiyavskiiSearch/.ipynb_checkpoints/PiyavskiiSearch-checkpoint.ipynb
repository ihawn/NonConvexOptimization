### Piyavskii Search is an alternative to line search and will find the true min along the search plane

### 2-d Objective

```julia
using Plots
```

```julia
SawTooth(x1, x2, z1, z2, L) = (z1 - z2)/2L + (x1 + x2)/2.0, (z1 + z2)/2 + L*(x1 - x2)/2

function Piyavskii(f, a, b, L, ϵ, m)
    fa = f(a); fb = f(b)
    x, y = SawTooth(a, b, fa, fb, L)

    x_list = [a, x, b]
    y_list = [fa, y, fb]
    X = [a, b]
    Y = [fa, fb]
    α_list = Float64[]
    β_list = Float64[]
    minpos = 0

    for i = 1:m
        minpos = argmin(y_list)
        y_list[minpos] = f(x_list[minpos])

        for j = 1:2
            pos1 = minpos - j%2
            pos2 = minpos + (j+1)%2
            x, y = SawTooth(x_list[pos1], x_list[pos2], y_list[pos1], y_list[pos2], L)
            append!(x_list, x); append!(y_list, y);
            append!(X, x)
            append!(Y, f(x))
        end

        p = sortperm(x_list); permute!(x_list, p); permute!(y_list, p)

        α = minimum(y_list)
        β, minpos = findmin(Y)
        append!(α_list, α); append!(β_list, β)


        if abs(β - α) <= ϵ        
            break
        end
    end

    x = X[minpos]
    y = Y[minpos]

    return x, y, x_list, y_list, α_list, β_list
end
```

```julia
x1 = [-1.9, (rand() - 0.5)*4, 0.0]
x2 = [1.9, (rand() - 0.5)*4, 0.0]

f(x,y) = x*sin(10*y) + y*cos(10*x) + 0.5*x^2 + 0.5*y^2

x1[3] = f(x1[1], x1[2])
x2[3] = f(x2[1], x2[2])
```

```julia
aa = -2:.02:2
bb = -2:.02:2

contour(aa, bb, f, fill=false, colorbar = false, dpi = 100)
plot!(size=(600,600), legend=:none)

plot!([x1[1], x2[1]], [x1[2], x2[2]], color=:blue, label = false)
scatter!([x1[1]], [x1[2]], color=:green, label = "x1")
scatter!([x2[1]], [x2[2]], color=:green, label = "x2")
```

```julia
projected_y(x) = (x1[2] - x2[2])/(x1[1] - x2[1]) * (x - x1[1]) + x1[2]
phi(x) = f(x, projected_y(x))
```

```julia
projectedMin = Piyavskii(phi, x1[1], x2[1], 12.0, 1e-5, 5000)
```

```julia
aa = -2:.02:2
bb = -2:.02:2

contour!(aa, bb, f, fill=false, colorbar = false, dpi = 100)
plot!(size=(600,600), legend=:none)

plot!([x1[1], x2[1]], [x1[2], x2[2]], color=:blue, label = false)
scatter!([x1[1]], [x1[2]], color=:green, label = "x1")
scatter!([x2[1]], [x2[2]], color=:green, label = "x2")
scatter!([projectedMin[1]], [projected_y(projectedMin[1])], color=:black)
```

```julia
projected_plotter(x) = f(x, projected_y(x))
```

```julia
plot(projected_plotter, xrange = (-2, 2), title = "Slice of 3-D Objective Intersecting a Plane", label = false)
scatter!([x1[1]], [x1[3]], color=:green, label = "Bounds")
scatter!([x2[1]], [x2[3]], color=:green, label = false)
scatter!([projectedMin[1]], [projected_plotter(projectedMin[1])], color=:black, label = "Min")
```

### N-d Objective

```julia
n = 20

x1 = (rand(n+1) .- 0.5)*10
x2 = (rand(n+1) .- 0.5)*10
x1[1] = -5
x2[1] = 5

f(x) = sum(sin(x[i])*sin(i*x[i]^2/pi)^20 for i in 1:n) #Michalewicz Function

x1[n] = f(x1[1:n])
x2[n] = f(x2[1:n])
```

```julia
projected_coord(x) = [(x1[i] - x2[i])/(x1[1] - x2[1]) * (x - x1[1]) + x1[i] for i in 2:n]
phi(x) = f([x; projected_coord(x)])
phi(1)
```

```julia
projectedMin = Piyavskii(phi, x1[1], x2[1], 40.0, 1e-5, 5000)
projected_plotter(x) = f([x; projected_coord(x)])
```

```julia
plot(projected_plotter, xrange = (-5, 5), title = "Slice of N-D Objective Intersecting (N-1)-D Hyperplane", label = false)
scatter!([projectedMin[1]], [projected_plotter(projectedMin[1])], color=:black, label = "Min")
```
